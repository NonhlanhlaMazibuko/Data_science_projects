```python
# importing necessary packages

import pandas as pd
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
"""
The SMILES representation of the four molecules(propane, ethene, cyclohexane 
and buckminsterfullerene) were taken from the ChemPub and converted into 
an RDKit molecule object.
Then information about each of the RDKit molecule objects was extracted using 
the Chem library functions to get number of atoms in the molecule objects,
the symbols and molecular weights of the atoms, and check for aromatic bonds
in the molecule object.
All the information extracted gives insights into understanding the size and
structure of the molecules, the elements present in the molecules. 
This information is very useful in understanding and predicting the chemical behaviour of 
molecules in various chemical contexts.
"""

# SMILES strings for the molecules
smiles_list = ["CCC", "C=C", "C1CCCCC1",
               "C12=C3C4=C5C6=C1C7=C8C9=C1C%10=C%11C(=C29)C3=C2C3=C4C4=C5C5=C9C6=C7C6=C7C8=C1C1=C8C%10=C%10C%11=C2C2=C3C3=C4C4=C5C5=C%11C%12=C(C6=C95)C7=C1C1=C%12C5=C%11C4=C3C3=C5C(=C81)C%10=C23"]

molecules = [Chem.MolFromSmiles(smile) for smile in smiles_list]

# Function to get the number of atoms
def get_num_atoms(mol):
    return mol.GetNumAtoms()

# Function to get atom symbols and their atomic weights
def get_atoms_info(mol):
    atoms_info = []
    for atom in mol.GetAtoms():
        symbol = atom.GetSymbol()
        atomic_weight = atom.GetMass()
        atoms_info.append((symbol, atomic_weight))
    return atoms_info

# Function to get the number of aromatic bonds
def get_number_of_aromatic_bonds(mol):
    aromatic_bonds = 0
    for bond in mol.GetBonds():
        if bond.GetIsAromatic():
            aromatic_bonds += 1
    return aromatic_bonds

# Loop through each molecule
for mol in molecules:
    # Get and print the number of atoms
    num_atoms = get_num_atoms(mol)
    print(f"Number of atoms in the molecule: {num_atoms}")

    # Get and print atom symbols and atomic weights
    atoms_info = get_atoms_info(mol)
    for symbol, atomic_weight in atoms_info:
        print(f"Symbol of atom: {symbol}, Atomic weight of atom: {atomic_weight}")

    # Get and print the number of aromatic bonds
    num_aromatic_bonds = get_number_of_aromatic_bonds(mol)
    print(f"Number of aromatic bonds: {num_aromatic_bonds}\n")
```

```python
"""
In the below section we will be visually showing the chemical structures
of the first six elements in the Daleney dataset imported in the above cell.
The drawing of the chemical structures of the molecules will be used to analyze
and compare the chemical behaviour and properties of the molecules.
Feature to be looked at are the geometric arrangements of the atoms in the molecules,
functional groups present, size of molecules, aromatic bonds and ring structures.
"""
```

```python
# Extract SMILES strings and Compound IDs
smiles_list = target_molecules['smiles'].tolist()
compound_ids = target_molecules['Compound ID'].tolist()

# Create RDKit molecule objects
molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
# Create a grid of molecule images
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()


# Draw each molecule
for i, (mol, cid) in enumerate(zip(molecules, compound_ids)):
    img = Draw.MolToImage(mol, size=(300, 300))
    axes[i].imshow(img)
    axes[i].set_title(f'ID: {cid}')
    axes[i].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
